```python
%run cards.ipynb
import itertools


# hand denotes a 4- or 5- card hand

# Variable Length Sub-Analysis

def is_15(hand_subset):
    value = 0
    for card in hand_subset:
        value += card.value()
    return value == 15

# Specific Length Sub-Analysis

def is_flush(hand_fourcard):
    x = return_suits(hand_fourcard)
    return x.count(x[0]) == len(x)
    
def is_pair(two_card_list):
    return two_card_list[0].rank == two_card_list[1].rank

# General Functions

def return_ranks(hand_subset):
    rank_list = []
    for card in hand_subset: rank_list.append(card.rank)
    return rank_list

def return_suits(hand_subset):
    suit_list = []
    for card in hand_subset: suit_list.append(card.suit)
    return suit_list



# Complete Hand Analysis Methods

def pair_count(hand_subset):
    if type(hand_subset[0]) == type(Card(1,1)):
        rank_list = return_ranks(hand_subset)
    else:
        rank_list = hand_subset
    
    pair_count = 0
    for pair in itertools.combinations(rank_list,2):
        if pair[0] == pair[1]:
            pair_count += 1
    
    return pair_count

def is_run(list_of_ranks):
    min_val = min(list_of_ranks)
    n = len(list_of_ranks)
    run = 1
    for i in range(n):
        if min_val + i not in list_of_ranks:
            run *= 0
    return run == 1

def straight_count(hand_subset):
    ranks = return_ranks(hand_subset)
    uniques = list(set(ranks))
    uniques.sort()
    
    str_len = 0
    str_tuple = None
    str_multiplier = 1
    
    if len(uniques) < 3:
        return 0
    else:
        levels = len(ranks)-len(uniques)

        while str_len == 0:
            for i in range(levels+1):
                for ntuple in itertools.combinations(uniques,len(ranks)-i):
                    card_list = list(ntuple)
                    if is_run(card_list):
                        str_len = len(ranks)-i
                        str_tuple = list(ntuple)
            break
    
    if str_tuple == None:
        return 0
    
    if len(str_tuple) < len(ranks):
        potential_pairs = []
        for rank in ranks:
            if (rank >= min(str_tuple) and rank <= max(str_tuple)):
                potential_pairs.append(rank)
        pairs = pair_count(potential_pairs)
        if pairs == 1:
            str_multiplier = 2
        elif pairs == 2:
            str_multiplier = 4
        elif pairs == 3:
            str_multiplier = 3
            
    return str_len*str_multiplier
            
    
def fifteen_count(hand_subset):
    hand_size = len(hand_subset)
    fifteens = 0
    for i in range(hand_size-1):
        for ntuple in itertools.combinations(hand_subset,2+i):
            if is_15(ntuple):
                fifteens += 1
    return fifteens

def right_jack_candidates(hand):
    jacks = []
    if 11 in return_ranks(hand):
        for card in hand:
            if card.rank == 11:
                jacks.append(card)
    return jacks

def score(hand,flip=None):
    # tags and counters
    score = 0
    flush = False
    
    # scoring not dependent on flip card
    if is_flush(hand) == True:
        score += 4
        flush = True
    
    # scoring dependent on flip card
    if flip != None: 
        hand.append(flip)
        
        jack_candidates = right_jack_candidates(hand)
        if jack_candidates != []:
            for card in jack_candidates:
                score += card.suit == flip.suit
        
        if flush: score += hand[0].suit == flip.suit    
    
    score += pair_count(hand)*2
    score += straight_count(hand)
    score += fifteen_count(hand)*2
    
    return(score)    
```
